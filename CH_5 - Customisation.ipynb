# Chapter 5.1 : Customisation of Scatter Plots

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
label_set = (  b'Iris-setosa',  b'Iris-versicolor',  b'Iris-virginica', )
```

```python
def read_label(label):  
    return label_set.index(label)
```

```python
data = np.loadtxt('Desktop/Matplotlib/iris.txt',delimiter = ',',converters = { 4 : read_label })
color_set = ('.20', '.55', '.90') 

color_list = [color_set[int(label)] for label in data[:,4]] 
              
plt.scatter(data[:,0], data[:,1], color = color_list) 
plt.show()
```

```python
color_set = ('r', 'b', 'm') 
color_list = [color_set[int(label)] for label in data[:,4]]
plt.scatter(data[:,0], data[:,1], color = color_list) 
plt.show()
```

# Chapter 5.2 : Customisation of BAR Plots

```python
import matplotlib.pyplot as plt
import numpy as np

data = [
    [5, 10 ,15, 19],
    [6, 12 ,14, 20],
]
x= 1+np.arange(4)
```

```python
men,women =np.array(data)
```

```python
plt.barh(x,-men,color ='red')
plt.barh(x,women,color ='blue')
plt.show()
```

## Colour Dependent on Values

```python
import matplotlib.pyplot as plt
import numpy as np

values = np.array([5, 10 ,15, 19])
x= 1+np.arange(4)
```

#### Color_list is created to map colour to specific interval

```python
color_set = ('.00', '.25', '.50', '.75') 
color_list = [color_set[(len(color_set) * val) // 20] for val in values] 
```

```python
plt.bar(np.arange(len(values)), values, color = color_list) 
plt.show() 
```

## HATCH , CROSS IN BAR

```python
plt.bar(np.arange(len(values)), values, color = color_list, hatch = '/') 
plt.show() 
```

```python
plt.bar(np.arange(len(values)), values, color = color_list, hatch = 'X') 
plt.show()
```

```python
plt.bar(np.arange(len(values)), values, color = color_list, hatch = '//') 
plt.show()
```

## TASK: COLORMAP

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
```

```python
current_palette = sns.color_palette("muted",n_colors=5)
cmap = ListedColormap(sns.color_palette(current_palette).as_hex())
```

```python
data1 = np.random.randn(500)
data2 = np.random.randn(500)
```

```python
colors = np.random.randint(0,5,500)
```

```python
plt.scatter(data1,data2,c=colors,cmap=cmap)
plt.colorbar()
plt.show()
```
