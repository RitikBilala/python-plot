# Chapter 6.1 : MARKERS  : Scatter Plot

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = np.random.standard_normal((100,2))
```

### Properties : TYPE: 'marker' and SIZE: 's'
<b>Possible marker values : +, ^, *, o and [0,9] and <br>
It can be (N,0,angle) or (N,1,angle) for regular polygon and star respectively<br>
Vertices can be given as input to form shapes</b>

```python
plt.scatter(data[:,0], data[:,1],c= 'g', edgecolor = '1',marker = 'o', s = 100) 
plt.show()
```

# Chapter 7.1: Custom Color Style : RC object

```python
import  matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
mpl.rc('lines',linewidth = 2)
mpl.rc('axes', facecolor = 'k', edgecolor ='g')
mpl.rc('xtick',color = 'w')
mpl.rc('ytick',color = 'w')
mpl.rc('text', color = 'g')
mpl.rc('figure', facecolor = 'k', edgecolor ='w')
```

```python
x = np.linspace(-10,10)
y = np.power(3,x)

plt.plot(x,y,c = 'green', markersize = 10,markevery = 5, marker = 'o',markerfacecolor='green',label='luck')
plt.legend(loc='best')
plt.show()
```

# Chapter 7.2: ANNOTATIONS

```python
import  matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10,10)
y = np.power(3,x)

plt.title('Profit vs Time')
plt.xlabel('time')
plt.ylabel('Profit')

plt.text(-5,10000, 'Great curve (-5,1000)')

plt.plot(x,y,c = 'green', markersize = 10,markevery = 5, marker = 'o',markerfacecolor='green',label='luck')
plt.show()
```

### Bounding Box params for text

```python
box = {'facecolor': '.75' , 'edgecolor': 'k' , 'boxstyle': 'round' }
```

```python
plt.text(-5,10000, 'Great curve' , bbox = box)

plt.plot(x,y,c = 'green', markersize = 10,markevery = 5, marker = 'o',markerfacecolor='green',label='luck')
plt.show()
```

### Lets add grids: <br>ls  = '--', 'dotted' etc <br>lw = linewidth

```python
plt.grid(True, lw = 2, ls = 'dotted', c = 'blue', which = 'both')

plt.plot(x,y,c = 'green', markersize = 10,markevery = 5, marker = 'o',markerfacecolor='green',label='luck')
plt.show()
```

### Adding ticks to the axis

```python
import matplotlib.ticker as ticker 

plt.grid(True, lw = 2, ls = 'dotted', c = 'blue', which = 'both')

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(2))

plt.plot(x,y,c = 'green', markersize = 10)
plt.show()
```

## Easy way to add ticks in BAR charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(5)
pos = np.array(range(len(x)))
name = ['mango','banana','orange','grapes','apple']
plt.bar(pos,x)
plt.xticks(pos,name,rotation = 30)

plt.show()
```
