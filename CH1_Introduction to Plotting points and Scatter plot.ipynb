### <b>Learn difference between np.arange() and np.linspace()</b>

## Task 2: Plot from file

Import Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Import Text File

```python
f = open("Desktop/Matplotlib/original.txt",r)
```

### Read Data from Text File

```python
X,Y = [],[]
for line in f:
    values = [float(s) for s in line.split()]
    X.append(values[0])
    Y.append(values[1])
```

### Learn how axis labels are customised:<br> we will explore more in coming chapter

```python
print(X,Y)
plt.plot(X,Y,color = 'red')

plt.xlabel("X from file", fontsize = 20, color = 'green')
plt.ylabel("Y from file", fontsize = 20, color = 'blue')
plt.show()
```

## Task 3: Scatter Plot using Pandas Dataframe

### Change colour and Size of dots on basis of data entries

```python
%matplotlib inline
##%matplotlib auto
##%matplotlib notebook
##%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw_data = {'first_name': ['Mike', 'Brad', 'Chan', 'Dee', 'Stan'], 
        'last_name': ['Miller', 'Smith', 'Man', 'Milner', 'Chin'], 
        'female': [0, 1, 1, 0, 1],
        'age': [42, 52, 36, 24, 73], 
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])
#try printing pd forlooking dataframe
df
```

### 's' sets size of the scatter: <br>We have kept size of scatter on basis of its value in data

```python
plt.scatter(df.preTestScore, df.postTestScore, s=df.age,c=df.female)
plt.show()
```
