# Chapter 3.3: Stacked Bar Graphs : Profit-Loss

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [[5, 10 ,15, 19],
        [6, 12 ,14, 20],
        [4, 9 ,13, 21]]
X= 1+np.arange(4);
A = np.array(data[0]);
B = np.array(data[1]);
C = np.array(data[2]);
```

```python
plt.bar(X, A, color = 'b') 
plt.bar(X, B, color = 'g', bottom = A) 
plt.bar(X, C, color = 'r', bottom = A + B)
plt.show()
```

```python
print(data[0] + data[1])
print(A + B)
#see why I used numpy and not list directly
```

# Chapter 4.1 :Pie Charts

shadow , startangle , autopct are options which customise  chart <ul><li>explode :creates space beween piece and center </li><li>autopct: prints percentage value with specified precision</li>

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Cookies', 'Jellybean', 'Milkshake', 'Cheesecake']
sizes = [38.4, 40.6, 20.7, 10.3]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="upper right")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## Color in Box Plot 

We're going to see how to use custom colors with box plots.

The plot function returns a dictionary.

The key of the dictionary is the name of the graphical element in the case of a box plot.

Such elements will be minion's flyers wiskers boxes and caps.

The values associated with each of those keys is a list of low level graphic primitives line shapes

and so on.

In our example on screen we iterate every graphic primitive.

That's part of the box plot and set it to the color black.

The same method allows us to render blocks plots with different color schemes.

Again if we use numbers in our color parameter we get a gray scale value and if we use letters that

map to colors in python we get colors for box plots.



```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
score = np.random.randn(150)
plot = plt.boxplot(score)
for name ,line_list in plot.items():
    for line in line_list:
        line.set_color('green')
part = plot.values()
plt.show()
```

Accessing plot.items and coloring each object with choice of colour is possible by the above method
plot has dictionary items of line to colour
